# **Feature Scaling**

```python
# separating the data and labels
X = data1.drop(columns = 'price', axis=1)
Y = data1['price']
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
pd.DataFrame(X)
```

# **TRAIN/VALIDATION/TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.15, random_state=42)
```

```python
X_train2, X_val, Y_train2, Y_val = train_test_split(X_train,Y_train, test_size = 0.15, random_state=42)
```

```python
pd.DataFrame(X_train2)
```

```python
Y_train2
```

# **Training the model**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

**Decision Tree**

```python
classifier = DecisionTreeRegressor()
```

```python
#training the Model
classifier.fit(X_train2, Y_train2)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
# accuracy score on the training data
X_train_prediction1 = classifier.predict(X_train2)
print('Score MAE for Decision Tree Training Data: {:.4f}' .format(mean_absolute_error(Y_train2, X_train_prediction1)))
print('Score MSE for Decision Tree Training Data: {:.4f}' .format(mean_squared_error(Y_train2, X_train_prediction1, squared = True)))
print('Score MAPE for Decision Tree Training Data:  {:.4f}' .format(mean_absolute_percentage_error(Y_train2, X_train_prediction1)))
print('Score R2 for Decision Tree Training Data: {:.4f}' .format(r2_score(Y_train2, X_train_prediction1)))
```

```python
# accuracy score on validation data
X_val_prediction1 = classifier.predict(X_val)
print('Score MAE for Decision Tree Validation data: {:.4f}' .format(mean_absolute_error(Y_val, X_val_prediction1)))
print('Score MSE for Decision Tree Validation data: {:.4f}' .format(mean_squared_error(Y_val, X_val_prediction1, squared = True)))
print('Score MAPE for Decision Tree Validation data:  {:.4f}' .format(mean_absolute_percentage_error(Y_val, X_val_prediction1)))
print('Score R2 for Decision Tree Validation data: {:.4f}' .format(r2_score(Y_val, X_val_prediction1)))
```

```python
# accuracy score on the testing data
X_test_prediction1 = classifier.predict(X_test)
print('Score MAE for Decision Tree Validation data: {:.4f}' .format(mean_absolute_error(Y_test, X_test_prediction1)))
print('Score MSE for Decision Tree Validation data: {:.4f}' .format(mean_squared_error(Y_test, X_test_prediction1, squared = True)))
print('Score MAPE for Decision Tree Validation data:  {:.4f}' .format(mean_absolute_percentage_error(Y_test, X_test_prediction1)))
print('Score R2 for Decision Tree Validation data: {:.4f}' .format(r2_score(Y_test, X_test_prediction1)))
```
